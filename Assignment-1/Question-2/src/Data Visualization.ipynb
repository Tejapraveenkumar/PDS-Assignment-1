```python
import pandas as pd
cleaned_data = pd.read_csv(r'C:\Users\tejap\Assignment-1\Question-2\Cleaned_Data\cleaned_StudentPerformance.csv')
```

# Visualization - 1

```python
# find the gender distribution among the dataset gathered.
import matplotlib.pyplot as plt
plt.bar(cleaned_data['gender'].value_counts().index, 
       cleaned_data['gender'].value_counts().values,
       color = ['gray', 'skyblue'])
plt.title('Gender Distribution')
plt.xticks(cleaned_data['gender'].value_counts().index,['Female','Male'])
```

# Visualization - 2

```python
import seaborn as sns

plt.scatter(cleaned_data['reading score'], cleaned_data['writing score'], c=cleaned_data['gender'])
plt.legend(['female'])
plt.xlabel('reading score')
plt.ylabel('writing score')
plt.title('reading score distribution vs writing score distribution')
```

# Visualization - 3

```python
figure, axis = plt.subplots(2,2)

sns.histplot(cleaned_data['math score'],ax=axis[0,0],label='math score')
sns.histplot(cleaned_data['reading score'],ax=axis[0,1],label='reading score')
sns.histplot(cleaned_data['writing score'],ax=axis[1,0],label='writing score')
sns.histplot(cleaned_data['Total score'],ax=axis[1,1],label='Total score')
plt.subplots_adjust(left=0.1,
                    bottom=0.1,
                    right=0.9,
                    top=0.9,
                    wspace=0.4,
                    hspace=0.4)
plt.suptitle('Frequency distribution of scores')
```

# Visualization - 4

```python
# different scores distribution based on gender
figure, axis = plt.subplots(2, 2)


sns.boxplot(x=cleaned_data['gender'],y=cleaned_data['math score'],ax=axis[0,0])
sns.boxplot(x=cleaned_data['gender'],y=cleaned_data['reading score'],ax=axis[0,1])
sns.boxplot(x=cleaned_data['gender'],y=cleaned_data['writing score'],ax=axis[1,0])
sns.boxplot(x=cleaned_data['gender'],y=cleaned_data['Total score'],ax=axis[1,1])
plt.subplots_adjust(left=0.1,
                    bottom=0.1,
                    right=0.9,
                    top=0.9,
                    wspace=0.4,
                    hspace=0.4)
plt.suptitle('Scores distribution by gender')
```

# Visualization - 5

```python
graph = sns.FacetGrid(cleaned_data, col ='gender',hue_kws={'color': ['g', 'b']}, hue ='test preparation course')
# map the above form facetgrid with some attributes
graph.map(plt.hist, 'race/ethnicity', bins = 15, color ='orange').add_legend()
```
