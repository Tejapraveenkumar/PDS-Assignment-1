```python
#Preprocessing the data to perform the analysis by changing categorical values of 'gender', 'test preparation course' attributes into numerical values

rawdata['gender'].replace(['female', 'male'],[0, 1], inplace=True)

rawdata['test preparation course'].replace(['none', 'completed'],[0, 1], inplace=True)

#Creating a new column total score by adding the values of the attributes 'math score', 'writing score' and 'reading score'
rawdata['Total score'] = rawdata['math score'] + rawdata['writing score'] + rawdata['reading score']
```

```python
#drop the rows if any missingvalues
cleaneddata=rawdata.dropna()
cleaneddata

#Save the cleaned data as csv file
cleaneddata.to_csv(r'C:\Users\tejap\Assignment-1\Question-2\Cleaned_Data/cleaned_StudentPerformance.csv')
print('successfully saved the data')
```
